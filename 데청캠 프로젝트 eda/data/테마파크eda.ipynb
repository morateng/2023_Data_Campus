```python
import pandas as pd
```

```python
df = pd.read_csv('../data/서울시유원지인허가정보.csv', encoding='cp949')
df
```

```python
df_cleaned = df[~df['영업상태명'].str.contains('폐업')]
df_cleaned
```

```python
df_cleaned['자치구'] = df_cleaned['지번주소'].str.split().str[1]
df_cleaned
```

```python
value_counts = df_cleaned['자치구'].value_counts()
value_counts
```

```python
new_dataframe = pd.DataFrame(value_counts, columns=['자치구'])
new_dataframe
```

```python
new_dataframe.rename(columns={'자치구':'테마파크 수'}, inplace=True)
new_dataframe
```
