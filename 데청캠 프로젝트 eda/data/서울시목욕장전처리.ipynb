```python
import pandas as pd
```

```python
df = pd.read_csv('../data/서울시 목욕장업 인허가 정보.csv', encoding='cp949')
df
```

```python
df.dropna(subset=['지번주소'], inplace=True)
df_seoul = df[df['지번주소'].str.contains('서울특별시')]
df_seoul['자치구'] = df_seoul['지번주소'].str.split().str[1]
df_seoul
```

```python
value_counts = df_seoul['자치구'].value_counts()
value_counts
```

```python
new_dataframe = pd.DataFrame(value_counts, columns=['자치구'])
new_dataframe
```

```python
new_dataframe.rename(columns={'자치구':'목욕탕 수'}, inplace=True)
new_dataframe
```

```python
new_dataframe.to_csv('../data/목욕탕 수.csv')
```
