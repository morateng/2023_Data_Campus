```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/시설수데이터+만족도.csv')
df.drop('자치구', axis=1, inplace=True)
df.drop('만족도(평균/5점만점)', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()   
df_scaled = pd.DataFrame(scaler.fit_transform(df))
df_scaled
```

```python
transposed_df = df_scaled.T
transposed_df
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=1)  # 주성분 개수 설정

# PCA 적용
transformed_data = pca.fit_transform(df_scaled)
```

```python
transformed_data.shape
```

```python
transformed_data
```

```python
abs_data = np.abs(transformed_data)

# MinMax 스케일링을 사용하여 0~5 범위로 조정
scaler = MinMaxScaler(feature_range=(0, 5))
scaled_data = scaler.fit_transform(abs_data)
```

```python
df = pd.DataFrame(scaled_data, columns=['여가시설점수 pca'])
```

```python
df
```

```python
data = pd.read_csv('../data/시설수데이터+만족도.csv')
df_자치구 = data.iloc[:,0]
df_자치구
```

```python
result_score = pd.concat([df_자치구, df], axis=1)
result_score
```

```python
result_score.to_csv('../data/시설점수_pca.csv')
```
