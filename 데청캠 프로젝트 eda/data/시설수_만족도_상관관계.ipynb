```python
import pandas as pd
```

```python
df = pd.read_csv('../data/시설수데이터+만족도.csv')
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
correlation_matrix = df.corr()
# 히트맵 그리기
plt.figure(figsize=(20, 20))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Variable Correlation Heatmap')
plt.show()
```

```python
df_문화예술시설 = df[['공연시설수', '전시시설수']]
df_문화예술시설
```

```python
from sklearn.preprocessing import StandardScaler

df_문화예술시설 = StandardScaler().fit_transform(df_문화예술시설)
df_문화예술시설
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=1)

pca.fit(df_문화예술시설)
문화예술시설_pca = pca.transform(df_문화예술시설)
```

```python
문화예술시설_df = pd.DataFrame(문화예술시설_pca, columns=['문화예술시설'])
문화예술시설_df
```

```python
df.columns
```

```python
df_스포츠시설 = df[['빙상장 수', '체육시설 수', '간이운동장 수', '체육도장 수', '골프/골프연습장 수', '당구장 수', '수영장 수', '체력단련장 수']]
df_스포츠시설 = StandardScaler().fit_transform(df_스포츠시설)
pca = PCA(n_components=1)

pca.fit(df_스포츠시설)
스포츠시설_pca = pca.transform(df_스포츠시설)
스포츠시설_df = pd.DataFrame(스포츠시설_pca, columns=['스포츠시설'])
스포츠시설_df
```

```python
df.columns
```

```python
df_관광시설 = df[['관광지 수', '동/식물원/휴양림 수', '유적지 수', '축제 수']]
df_관광시설 = StandardScaler().fit_transform(df_관광시설)
pca = PCA(n_components=1)

pca.fit(df_관광시설)
관광시설_pca = pca.transform(df_관광시설)
관광시설_df = pd.DataFrame(스포츠시설_pca, columns=['관광시설'])
관광시설_df
```

```python
df.columns
```

```python
df_취미오락시설 = df[['노래방시설수', '오락실시설수', '영화관 수', '도서관 수']]
df_취미오락시설 = StandardScaler().fit_transform(df_취미오락시설)
pca = PCA(n_components=1)

pca.fit(df_취미오락시설)
취미오락시설_pca = pca.transform(df_취미오락시설)
취미오락시설_df = pd.DataFrame(취미오락시설_pca, columns=['취미오락시설'])
취미오락시설_df
```

```python
df.columns
```

```python
df_휴식시설 = df[['호텔수', '공원 수', '목욕탕 수']]
df_휴식시설 = StandardScaler().fit_transform(df_휴식시설)
pca = PCA(n_components=1)

pca.fit(df_휴식시설)
휴식시설_pca = pca.transform(df_휴식시설)
휴식시설_df = pd.DataFrame(휴식시설_pca, columns=['휴식시설'])
휴식시설_df
```

```python
df.columns
```

```python
df_소비시설  = df[['편의시설수', '음식점 수', '패션/미용소비 수', '대형소비시설 수']]
df_소비시설 = StandardScaler().fit_transform(df_소비시설)
pca = PCA(n_components=1)

pca.fit(df_소비시설)
소비시설_pca = pca.transform(df_소비시설)
소비시설_df = pd.DataFrame(소비시설_pca, columns=['소비시설'])
소비시설_df
```

```python
df.columns
```

```python
df_노인시설 = df[['노인복지관 시설수', '경로당 시설수', '노인교실 시설수']]
df_노인시설 = StandardScaler().fit_transform(df_노인시설)
pca = PCA(n_components=1)

pca.fit(df_노인시설)
노인시설_pca = pca.transform(df_노인시설)
노인시설_df = pd.DataFrame(노인시설_pca, columns=['노인시설'])
노인시설_df
```

```python
df_시설수_대분류_pca = pd.concat([문화예술시설_df, 스포츠시설_df, 관광시설_df, 취미오락시설_df, 휴식시설_df, 소비시설_df, 노인시설_df], axis=1)
df_시설수_대분류_pca
```

```python
df_시설수_대분류_pca.to_csv('../data/시설수_대분류_pca.csv')
```
