```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
x_df = pd.read_csv('../data/자치구별여가시설수.csv')
y_df = pd.read_csv('../data/자치구별 만족도.csv')
```

```python
x_df.sort_values(by='자치구', inplace=True)
x_df.set_index('자치구', inplace=True)
x_df
```

```python
y_df.rename(columns={'자치구별': '자치구'}, inplace=True)
y_df.set_index('자치구', inplace=True)
y_df
```

```python
data = pd.concat([x_df, y_df], axis=1)
data
```

```python
data.to_csv('../data/시설수데이터+만족도.csv')
```
