```python
import pandas as pd
```

```python
df = pd.read_csv('../data/문화예술체육소비시군구별시설수.csv')
```

```python
df_seoul = df[df['SIGNGU_NM'].str.contains('서울특별시')]
df_seoul['자치구'] = df_seoul['SIGNGU_NM'].str.split().str[1]
df_seoul
```

```python
df_seoul['MLSFC_NM'].unique()
```

```python
df_seoul_공연시설 = df_seoul[df_seoul['MLSFC_NM'].str.contains('공연시설')]
df_seoul_공연시설
```

```python
df_seoul_공연시설.rename(columns={'FCLTY_CO':'공연시설수'})
```

```python
df_seoul_공연시설.drop(['LCLAS_NM', 'MLSFC_NM', 'SIGNGU_NM'], axis=1, inplace=True)
```

```python
df_seoul_공연시설.rename(columns={'FCLTY_CO':'공연시설수'}, inplace=True)
```

```python
df_seoul_공연시설 = df_seoul_공연시설.reindex(['자치구', '공연시설수'], axis=1)
```

```python
df_seoul_공연시설
```

```python
df_seoul_전시시설 = df_seoul[df_seoul['MLSFC_NM'].str.contains('전시시설')]
df_seoul_전시시설.rename(columns={'FCLTY_CO':'전시시설수'}, inplace=True)
df_seoul_전시시설.drop(['LCLAS_NM', 'MLSFC_NM', 'SIGNGU_NM'], axis=1, inplace=True)
df_seoul_전시시설 = df_seoul_전시시설.reindex(['자치구', '전시시설수'], axis=1)
df_seoul_전시시설
```

```python
df_seoul_지역문화시설 = df_seoul[df_seoul['MLSFC_NM'].str.contains('지역문화시설')]
df_seoul_지역문화시설.rename(columns={'FCLTY_CO':'지역문화시설수'}, inplace=True)
df_seoul_지역문화시설.drop(['LCLAS_NM', 'MLSFC_NM', 'SIGNGU_NM'], axis=1, inplace=True)
df_seoul_지역문화시설 = df_seoul_지역문화시설.reindex(['자치구', '지역문화시설수'], axis=1)
df_seoul_지역문화시설
```

```python
df_seoul_학교_교육기관 = df_seoul[df_seoul['MLSFC_NM'].str.contains('학교/교육기관')]
df_seoul_학교_교육기관.rename(columns={'FCLTY_CO':'학교/교육기관 수'}, inplace=True)
df_seoul_학교_교육기관.drop(['LCLAS_NM', 'MLSFC_NM', 'SIGNGU_NM'], axis=1, inplace=True)
df_seoul_학교_교육기관 = df_seoul_학교_교육기관.reindex(['자치구', '학교/교육기관 수'], axis=1)
df_seoul_학교_교육기관
```

```python
df_seoul_공원 = df_seoul[df_seoul['MLSFC_NM'].str.contains('공원')]
df_seoul_공원.rename(columns={'FCLTY_CO':'공원 수'}, inplace=True)
df_seoul_공원.drop(['LCLAS_NM', 'MLSFC_NM', 'SIGNGU_NM'], axis=1, inplace=True)
df_seoul_공원 = df_seoul_공원.reindex(['자치구', '공원 수'], axis=1)
df_seoul_공원
```

```python
df_seoul_관광지 = df_seoul[df_seoul['MLSFC_NM'].str.contains('관광지')]
df_seoul_관광지.rename(columns={'FCLTY_CO':'관광지 수'}, inplace=True)
df_seoul_관광지.drop(['LCLAS_NM', 'MLSFC_NM', 'SIGNGU_NM'], axis=1, inplace=True)
df_seoul_관광지 = df_seoul_관광지.reindex(['자치구', '관광지 수'], axis=1)
df_seoul_관광지
```

```python
df_seoul_동식물원_휴양림 = df_seoul[df_seoul['MLSFC_NM'].str.contains('동/식물원/휴양림')]
df_seoul_동식물원_휴양림.rename(columns={'FCLTY_CO':'동/식물원/휴양림 수'}, inplace=True)
df_seoul_동식물원_휴양림.drop(['LCLAS_NM', 'MLSFC_NM', 'SIGNGU_NM'], axis=1, inplace=True)
df_seoul_동식물원_휴양림 = df_seoul_동식물원_휴양림.reindex(['자치구', '동/식물원/휴양림 수'], axis=1)
df_seoul_동식물원_휴양림
```

```python
df_seoul_음식소비 = df_seoul[df_seoul['MLSFC_NM'].str.contains('음식소비')]
df_seoul_음식소비.rename(columns={'FCLTY_CO':'음식점 수'}, inplace=True)
df_seoul_음식소비.drop(['LCLAS_NM', 'MLSFC_NM', 'SIGNGU_NM'], axis=1, inplace=True)
df_seoul_음식소비 = df_seoul_음식소비.reindex(['자치구', '음식점 수'], axis=1)
df_seoul_음식소비
```

```python
df_seoul_패션미용소비 = df_seoul[df_seoul['MLSFC_NM'].str.contains('패션/미용소비')]
df_seoul_패션미용소비.rename(columns={'FCLTY_CO':'패션/미용소비 수'}, inplace=True)
df_seoul_패션미용소비.drop(['LCLAS_NM', 'MLSFC_NM', 'SIGNGU_NM'], axis=1, inplace=True)
df_seoul_패션미용소비 = df_seoul_패션미용소비.reindex(['자치구', '패션/미용소비 수'], axis=1)
df_seoul_패션미용소비
```

```python
df_seoul_대형소비시설 = df_seoul[df_seoul['MLSFC_NM'].str.contains('대형소비시설')]
df_seoul_대형소비시설.rename(columns={'FCLTY_CO':'대형소비시설 수'}, inplace=True)
df_seoul_대형소비시설.drop(['LCLAS_NM', 'MLSFC_NM', 'SIGNGU_NM'], axis=1, inplace=True)
df_seoul_대형소비시설 = df_seoul_대형소비시설.reindex(['자치구', '대형소비시설 수'], axis=1)
df_seoul_대형소비시설
```

```python
df_seoul_공연시설.set_index('자치구', inplace=True)
df_seoul_공연시설
```

```python
df_seoul_전시시설.set_index('자치구', inplace=True)
df_seoul_지역문화시설.set_index('자치구', inplace=True)
df_seoul_관광지.set_index('자치구', inplace=True)
df_seoul_공원.set_index('자치구', inplace=True)
df_seoul_동식물원_휴양림.set_index('자치구', inplace=True)
df_seoul_음식소비.set_index('자치구', inplace=True)
df_seoul_패션미용소비.set_index('자치구', inplace=True)
df_seoul_대형소비시설.set_index('자치구', inplace=True)
```

```python
result = pd.concat([df_seoul_공연시설, df_seoul_전시시설, df_seoul_관광지, df_seoul_공원, df_seoul_동식물원_휴양림, df_seoul_음식소비, df_seoul_패션미용소비,
                    df_seoul_대형소비시설], axis=1)
result
```

```python
result.to_csv('../data/여가시설수전처리.csv')
```
