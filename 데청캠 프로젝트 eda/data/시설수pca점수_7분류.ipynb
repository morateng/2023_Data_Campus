```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/자치구별여가시설수_4차.csv')
df.head()
```

```python
df.sort_values(by='자치구', inplace=True)
```

```python
df.head()
```

```python
df.drop('자치구', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()   
df_scaled = pd.DataFrame(scaler.fit_transform(df))
df_scaled
```

```python
from sklearn.decomposition import PCA

pca = PCA()  # 주성분 개수 설정

# PCA 적용
transformed_data = pca.fit_transform(df_scaled)
```

```python
explained_var_ratio = pca.explained_variance_ratio_

# explain_ratio를 가중치로 주어 주성분의 중요도를 표현
weighted_var_ratio = explained_var_ratio / np.sum(explained_var_ratio)
```

```python
print("PCA-based Weights:")
for i, weight in enumerate(weighted_var_ratio):
    print(f"Variable {i + 1}: {weight:.3f}")
```

```python
indicator_scores = []
for i in range(len(df_scaled)):
    score = np.dot(df_scaled.iloc[i], weighted_var_ratio)    
    indicator_scores.append(score)

# 결과를 데이터프레임으로 변환
indicator_scores_df = pd.DataFrame({'Indicator_Score': indicator_scores})

# 결과 출력
indicator_scores_df
```
