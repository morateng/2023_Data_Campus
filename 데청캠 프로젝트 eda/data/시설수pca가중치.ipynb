```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/시설수데이터+만족도.csv')
df.drop('자치구', axis=1, inplace=True)
df.drop('만족도(평균/5점만점)', axis=1, inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler

df_scaled = StandardScaler().fit_transform(df)
df_scaled
```

```python
from sklearn.decomposition import PCA

# Perform PCA
pca = PCA()
principal_components = pca.fit_transform(df_scaled)

# Calculate explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Calculate weights based on explained variance ratio
weights = explained_variance_ratio / np.sum(explained_variance_ratio)

# Print the weights
print("PCA-based Weights:")
for i, weight in enumerate(weights):
    print(f"Variable {i + 1}: {weight:.50f}")
```

```python
from sklearn.decomposition import PCA



# Calculate the correlation matrix
#correlation_matrix = df_scaled.corr()

# Perform PCA
pca = PCA()
pca.fit(df_scaled)

# 주성분(PC)과 그에 대응하는 가중치 출력
for i, pc in enumerate(pca.components_):
    print(f"Principal Component {i + 1} weights:")
    for j, weight in enumerate(pc):
        print(f"Variable {j + 1}: {weight}")
```
