```python
import pandas as pd
```

```python
df = pd.read_csv('../data/서울시문화행사정보.csv', encoding='cp949')
df
```

```python
value_counts = df['자치구'].value_counts()
value_counts
```

```python
new_dataframe = pd.DataFrame(value_counts, columns=['자치구'])
new_dataframe
```

```python
new_dataframe.rename(columns={'자치구':'축제 수'}, inplace=True)
new_dataframe
```

```python
new_dataframe.to_csv('../data/지역축제 수.csv')
```
